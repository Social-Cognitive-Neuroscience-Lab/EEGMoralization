# Re-analysis of ERP data with Moral Conviction as parametric modulator

Extract each individual trial, group by moral conviction rating, then re-run ERP analysis within a multi-level model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as pl
import pickle
import mne
from pathlib import Path
```

```python
# change for each local context
base_dir = Path('D:/PM_Moral/export/') #Path('/Volumes/Storage/PM_Moral/export/')  
stim_dir = Path('D:/PM_Moral/stim_files/') #Path('/Volumes/Storage/PM_Moral/stim_files/') 
repo_dir = Path('C:/Users/Keith Yoder/repos/PoliticalMoralization') #Path('/Users/keithyoder/repos/PoliticalMoralization/')
```

```python
## Set up defaults
subs = pd.read_csv(base_dir / 'eeg_subs.csv',
                   header=None, names=['SID'])
chans = pd.read_csv(repo_dir / 'EasyCap_64chan.csv')

sub_labels = subs.SID.values.tolist()
chan_labels = chans.ChanID.values.tolist()



# read in all subject ratings
ratings_all = pd.read_csv(stim_dir / 'Ratings.csv')


# ERP time windows
time_dict = {
    'iss': {
        'epn': [100, 200],
        'mfn': [300, 500],
        'lpp': [600, 800],
    },
    'soc' : {
        'epn': [125, 225],
        'mfn': [200, 300],
        'lpp': [300, 500]
    },
    'pic' : {
        'epn': [100, 200],
        'mfn': [200, 400],
        'lpp': [400, 600]
    }
}

# electrode clusters
clust_dict = {
    'front': ['Fz', 'F1', 'F2', 'FCz', 'FC1', 'FC2', 'Cz', 'C1', 'C2'],
    'post' : ['Pz', 'P1', 'P2', 'POz', 'PO3', 'PO4', 'Oz', 'O1', 'O2']
}

# baseline for ERP epochs (ms)
bl = 200

# artifact criteria
reject_criteria = dict(eeg = 200e-6) # 100  µV
flat_criteria = dict(eeg = 0.5e-6) # 1 µV

# prep list of all possible codes
issue_vals = []
for s in range(7):
    for m in range(9):
        issue_vals.append( (s+1)*10 + m+1 )
soc_vals = []
for i in [100, 200, 300]:
    soc_vals = soc_vals + [x + i for x in issue_vals]
pic_vals = []
for i in [1000, 2000]:
    pic_vals = pic_vals + [x + i for x in soc_vals]
    
def import_eeg(subject, **kwargs):
    """
    Read in the EDF for the given subject.
    
    Parameters
    ----------
    subject : char
        The subject to read in
        
    Returns
    -------
    raw
        mne RawEDF file
    """
    
    raw = mne.io.read_raw_edf(base_dir / f'{subject}_Protest.edf', **kwargs)
    raw.info['lowpass'] = 0.1
    raw.info['highpass'] = 50
    raw.set_montage('easycap-M1')
    
    return raw

def import_markers(subject):
    """
    Read in the *.Marker file for the given subject.
    
    Parameters
    ----------
    subject : char
        The subject to read in
        
    Returns
    -------
    markers
        pandas data.frame of markers
    """
    
    d = pd.read_csv(base_dir / f'{subject}_Protest.Markers', skiprows=1,
                    skipinitialspace=True)
    d = d[ d['Description'] != 'Blink' ].reset_index(drop=True)
    
    return d

def import_stimorder(subject):
    """
    Read in the stimulus order for the given subject.
    
    Parameters
    ----------
    subject : char
        The subject to read in stimulus order
        
    Returns
    -------
    stims
        pandas data.frame with stimulus order
    """
    
    sub_runs = []
    for run in range(1,7):
        d = pd.read_csv(stim_dir / f'{subject}_run{run}.txt', delimiter='\t')
        sub_runs.append(d)
        
    stims = pd.concat(sub_runs)
    return stims['Issue']

def import_subeeg(subject, **kwargs):
    """    
    Read in subject's EEG data and create the events and event_dict.
    
    Parameters
    ----------
    subject : char
        The subject to be processed
        
    Returns
    -------
    eeg
        the EEG data as a mne.io.edf.edf.RawEDF
    events
        numpy.ndarray of events
    event_dict
        dict of events
    """
    
    eeg = import_eeg(subject, **kwargs)
    ann = mne.read_annotations(f'./markers/{subject}.vmrk')
    eeg.set_annotations(ann)
    events, event_dict = mne.events_from_annotations(eeg, parse_annot)
    
    return (eeg, events, event_dict)

def parse_markers(markers):
    """
    Recode the stimulus markers to have unique codes except for issue onsets.
    All issue onsets are coded as 42 for later scoring by score_markers
    
    Parameters
    ----------
    markers : pandas data.frame
        The pandas representation of subject markers (from import_markers)
        
    Returns
    -------
    markers
        pandas data.frame with rescored stimulus codes
    """
    
    trial_dict = {
        'ProtestStart': 2,
        'SocLow'  : 100,
        'SocMed'  : 200,
        'SocHigh' : 300,
        'PhotoOppose' : 1000,
        'PhotoSupport': 2000,
        'Resp1' : 401,
        'Resp2' : 402,
        'Resp3' : 403,
        'Resp4' : 404,
        'Resp5' : 405,
        'Resp6' : 406,
        'Resp7' : 407,
        'Moral_Support' : 42,
        'Moral_Neutral' : 42,
        'Moral_Oppose'  : 42,
        'Nonmoral_Support' : 42,
        'Nonmoral_Neutral' : 42,
        'Nonmoral_Oppose'  : 42
    }
    
    markers['Code'] = markers['Description']
    markers = markers.replace( {'Code': trial_dict} )
    
    return markers

def score_markers(markers, stimorder, ratings):
    """
    Replace all issue onsets with unique int codes to indicate
    subject's Support and MoralConviction for that issue
    
    Parameters
    ----------
    markers : pandas data.frame
        all participant markers
    stimorder : pandas Series
        list of issues in order presented during EEG task
    ratings : pandas data.frame
        participant Support and Moral Conviction ratings for each issue
        
    Returns
    -------
    markers
        pandas data.frame issue codes replaced with support-mc values
        
    Note
    ----
    Unique issue codes are generated using a two-digit integer
        first-digit is the support for the issue
        second-digit is the moral conviction for the issue
    """
    
    stim_rows = markers[ markers['Code'] == 42 ].index
    sub_ratings = ratings[ ['IssueText', 'MC', 'Support'] ]

    soc_types = ['SocLow', 'SocMed', 'SocHigh']
    pic_types = ['PhotoSupport', 'PhotoOppose']
    
    n_iss = len(stim_rows)
    n_soc = 0
    n_pic = 0
    
    n_bad = 0
    bad_issues = []
    
    for i in range(n_iss):
        stim_issue = stimorder.iloc[i]
        issue_idx = stim_rows[i]
        soc_idx = issue_idx + 1
        pic_idx = issue_idx + 2
        # in case things blow up
        if soc_idx >= markers.shape[0]:
            soc_idx = markers.shape[0] - 1
        if pic_idx >= markers.shape[0]:
            pic_idx = markers.shape[0] - 1
        
        soc_val = []
        pic_val = []
            
        stim_rating = sub_ratings[sub_ratings['IssueText'] == stim_issue][['MC','Support']]
        if stim_rating.empty:
            n_bad += 1
            bad_issues.append(stim_issue)
            #print(f'Could not find ratings for Issue "{stim_issue}"')
            new_val = int(-999)
        else:        
            support = stim_rating['Support'].values[0]
            moral = stim_rating['MC'].values[0]
            
            new_val = int( support*10 + (moral*2 - 1) )
            
            # check for next social and photo values
            soc_r = markers.iloc[soc_idx, markers.columns.get_loc('Description')] 
            pic_r = markers.iloc[pic_idx, markers.columns.get_loc('Description')]
            
            if soc_r in soc_types:
                soc_val = markers.iloc[soc_idx, markers.columns.get_loc('Code')]
                n_soc += 1
            else:
                print(f'Problem for stim {i}. Social value at {soc_idx} is {soc_r}.')
            # Rescore Social markers to be
            # Minority= 100: participants view is in the minority
            # Majority= 300: participants view is in the majority
            # Neutral=  200:  participant or social support was neutral
            if support == 4 or soc_r == 'SocMed':
                soc_val = 200 # neutral
            elif soc_r == "SocHigh":
                if support > 4:
                    soc_val = 300
                else: # support is low
                    soc_val = 100
            elif soc_r == "SocLow":
                if support < 4:
                    soc_val = 300
                else:
                    soc_val = 100
            else:
                print(f'\n\n\n\nBIG PROBLEM: Unhandled social rescoring\n\tsoc_r: {soc_r}  support: {support}\n\n\n')
            
            
            if pic_r in pic_types:
                pic_val = markers.iloc[pic_idx, markers.columns.get_loc('Code')]
                n_pic += 1
            else:
                print(f'Problem for stim {i}. Photo value at {pic_idx} value is {pic_r}.')
            
            # Rescore Pic markers to be
            # 
            # Disagree = 1000: participants agree with protestors
            # Unclear  = 2000: participants were neutral about issue
            # Agree    = 3000: participants disagree with the protestors
            if support == 4:
                pic_val = 2000
            elif pic_r == 'PhotoSupport':
                if support > 4:
                    pic_val = 3000
                else:
                    pic_val = 1000
            else: # oppose
                if support < 4: # also oppose
                    pic_val = 3000
                else:
                    pic_val = 1000
            
        # rescore markers
        markers.iloc[issue_idx, markers.columns.get_loc('Code')] = new_val
        if soc_val:
            markers.iloc[soc_idx, markers.columns.get_loc('Code')] = int(new_val + soc_val)
            if pic_val:
                markers.iloc[pic_idx, markers.columns.get_loc('Code')] = int(new_val + soc_val + pic_val)
    
    if not (n_soc == n_iss and n_pic == n_iss):
        print(f'\tMismatch between issues [{n_iss}], social support [{n_soc}], and photos [{n_pic}]')
    if n_bad > 0:
        print(f'\t{n_bad} issues could not be identified')
        
    # set new segment code to -999
    markers.iloc[0,5] = -999
        
    return markers

def process_subject(subject, rating_list=ratings_all, clobber=False):
    """
    Read in subject markers and stimulus order, then recode all issue
    markers to reflect subject's support and moral convictions about 
    that issue. Save subject stim_order, ratings, and rescored markers
    to a pickled file.
    
    Parameters
    ----------
    subject : char
        The subject to be processed
    (optional)
    rating_list : pandas DataFrame
        list of all ratings from all participants (default: ratings_all from setup)
    clobber : bool
        whether to overwrite out_file if it already exists (default: false)
        
    Returns
    -------
    stim
        pandas Series with stimulus order
    ratings
        pandas data.frame with participant issue ratings
    markers_scored
        pandas data.frame with re-scored markers
    """
    
    out_file = base_dir / f'{subject}_remarked.p'
    out_markers = Path(f'./markers/{subject}.vmrk')
    if out_file.exists() and not clobber:
        if out_markers.exists() and not clobber:
            return
        else:
            stim, ratings, markers_scored = pickle.load(open(out_markers,'rb'))
            write_markers(subject, markers_scored)
    else:
    
        print(f'Processing subject {subject}.')
        #eeg = import_eeg(subject)
        markers = parse_markers( import_markers(subject) )
        stim = import_stimorder(subject)
        ratings = rating_list[ rating_list['SID'] == subject]
        markers_scored = score_markers(markers.copy(), stim, ratings)
        processed = (stim, ratings, markers_scored)
        pickle.dump(processed, open(out_file, 'wb'))
        write_markers(subject, markers_scored, clobber=clobber)
                
def import_processed(subject):
    """    
    Read in stim_order, pre-test ratings, and rescored markers for the given subject.
    
    Parameters
    ----------
    subject : str
        the subject to import
        
    Returns
    -------
    stim_order
        the order of stimuli presented to the participant
    ratings
        the support and moral conviction ratings from the pretest
    markers_scored
        the rescored markers
    """
    
    dat = pickle.load(open(base_dir / f'{subject}_remarked.p', 'rb'))
    return(dat)

def write_markers(subject, markers, clobber=False):
    """
    Write out the subject's markers to a *.vmrk file for importing via mne.
    
    Parameters
    ----------
    subject : char
        The subject to be written (must be same as EEG file)
    markers : pandas data.frame
        The markers to be written
    (optional)
    clobber : bool
        Whether to overwrite marker file if it already exists
    """
    
    out_file = Path(f'./markers/{subject}.vmrk')
    if out_file.exists() and not clobber:
        return
    
    header = '\n'.join([
    'Brain Vision Data Exchange Marker File, Version 1.0\n\n[Common Infos]\nCodepage=UTF-8\nDataFile=JEMSUBJECT.eeg\n\n[Marker Infos]',
    '; Each entry: Mk<Marker number>=<Type>,<Description>,<Position in data points>',
    '; <Size in data points>, <Channel number (0 = marker is related to all channels)>',
    '; Fields are delimited by commas, some fields might be omitted (empty).',
    '; Commas in type or description text are coded as "\\1".'
    ])
    header = header.replace('JEMSUBJECT',subject)
    
    out = open(out_file, 'w')
    out.write(header)
    
    for ii in markers.index:
        mark = markers.iloc[ii,]
        code = mark['Code']
        if isinstance(code, str):
            # skip unprocessed codes
            pass
        else:
            if np.isnan(mark['Code']):
                code = 'New Segment'
            pos = mark['Position']*8 # mne refused to believe these are 250Hz and not 2kHz
            out.write(f"\nMk{ii+1}=Stimulus,{code},{pos},1,0")
    out.close()
    print(f'New markers written to {out_file}')
    
def parse_annot(a):
    """
    Helper function for creating mne events from imported markers.
    
    Removes leading 'Stimulus/' from each event so only event code is left
    
    Parameters
    ----------
    a : char
        stimulus identifier
        
    Returns
    -------
    code
        the integer code (or None if not an integer event)
    """
    
    code = a.replace('Stimulus/','')
    if 'S' in code:
        return None
    else:
        return int(float(code))
    

def parse_issue(code, bin_moral=False):
    """    
    Convert a two-digit issue code into support and moral conviction scores.
    
    Parameters
    ----------
    code : int
        The code to parse
        
    Returns
    -------
    support
        the support (int [1..8])
    moral
        the moral conviciton (int [1..10]
    """
    
    support = int(code/10)
    moral = code % 10
    
    if bin_moral:
        if moral < 3:
            moral = 1
        elif moral < 5:
            moral = 2
        elif moral == 5:
            moral = 3
        elif moral < 8:
            moral = 4
        else:
            moral = 5
            
    return (support, moral)

def parse_soc(code, bin_moral=False):
    """    
    Convert a three-digit issue code into support and moral conviction scores.
    
    Parameters
    ----------
    code : int
        The code to parse
        
    Returns
    -------
    soc
        the level of social support (1=Low, 2=Med, 3=High)
    support
        the support (int [1..8])
    moral
        the moral conviciton (int [1..10]
    """
    
    soc = int(code/100)
    tmp = code - soc*100
    support = int(tmp/10)
    moral = tmp % 10
    
    if bin_moral:
        if moral < 3:
            moral = 1
        elif moral < 5:
            moral = 2
        elif moral == 5:
            moral = 3
        elif moral < 8:
            moral = 4
        else:
            moral = 5
            
    return (soc, support, moral)

def parse_pic(code, bin_moral=False):
    """    
    Convert a four-digit photo code into support and moral conviction scores.
    
    Parameters
    ----------
    code : int
        The code to parse
        
    Returns
    -------
    position
        the protestors' position (1=Oppose, 2=Support)
    soc
        the level of social support (1=Low, 2=Med, 3=High)
    support
        the support (int [1..8])
    moral
        the moral conviciton (int [1..10]
    """
    
    pos = int(code / 1000)
    tmp = code - pos*1000
    soc = int(tmp/100)
    tmp = tmp % 100
    support = int(tmp/10)
    moral = tmp % 10
    
    if bin_moral:
        if moral < 3:
            moral = 1
        elif moral < 5:
            moral = 2
        elif moral == 5:
            moral = 3
        elif moral < 8:
            moral = 4
        else:
            moral = 5
    return (pos, soc, support, moral)

# convert to pnts
def ms2pnt(ms, srate=250):
    """    
    Convert from milliseconds to sample points.
    
    Parameters
    ----------
    ms : int
        value (in msec) to convert to sample points
    srate : int
        the sampling rate (default: 250 Hz)
        
    Returns
    -------
    pnt
        the sample point corresponding to the given ms
    """
    
    p_time = srate/1000
    pnt = int(ms*p_time)
    return(pnt)

def export_issue(subject, eeg, events, issue_codes, clust_dict=clust_dict,
                 reject_criteria=reject_criteria, flat_criteria=flat_criteria,
                bin_moral=False):
    """    
    Calculate the mean voltage for each channel within the EPN, MFN, and LPP 
    time windows, separately for each distinct moral conviction rating.
    
    Parameters
    ----------
    subject : char
        Subject id
    eeg : pandas DataFrame
        subject's eeg from import_subeeg
    events : pandas DataFrame
        subjects event structure from import_subeeg
    issue_codes : list
        all codes for issues that exist for this subject
        
    Returns
    -------
    iss_df : pandas DataFrame
        collected data written to file
    """
    erps = ['epn', 'mfn', 'lpp']
    clusters = ['front', 'post']
    iss_sup = []
    iss_mor = []
    iss_codes = []
    for iss in issue_codes:
        iss_codes.append(iss)
        s, m = parse_issue(iss)
        iss_sup.append(s)
        iss_mor.append(m)
    
    sub_df = pd.DataFrame({'iss_id': iss_codes, 
                           'support' : iss_sup, 
                           'moral' : iss_mor})
    
    if bin_moral:
        n_vals = sub_df['moral'].unique().shape[0]
        cases = [
            (sub_df["moral"].lt(3)),
            (sub_df["moral"].ge(3) & sub_df["moral"].lt(5)),
            (sub_df["moral"] == 5),
            (sub_df["moral"].gt(5) & sub_df["moral"].le(7)),
            (sub_df["moral"].gt(7)),
        ]
        bins = [1, 2, 3, 4, 5]

        sub_df["moral_bin"] = np.select(cases, bins)
        sub_df['moral'] = sub_df['moral_bin']
        n_vals = sub_df['moral'].unique().shape[0]
    
    
    iss_df = pd.DataFrame(
        columns=['Subject','Channel','Cluster','ERP','Support','Moral','uV']
    )

    dfs = []

    for clust in clusters:
        channels = clust_dict[clust]

        for ui in sub_df['moral'].unique():
            code_vals = sub_df[sub_df['moral'] == ui]['iss_id'].values
            s, m = parse_issue(code_vals[0], bin_moral)
            epoch = mne.Epochs(eeg, events, event_id = code_vals.tolist(), tmin=-0.2, tmax=1,
                               picks = channels, reject=reject_criteria,
                               flat=flat_criteria, reject_by_annotation=False,
                               preload=True, verbose='WARNING')

            for erp in erps:
                #print(f'Now working with erp {erp}')
                win = [ms2pnt(x+bl) for x in time_dict['iss'][erp]]
                # n_seg x n_chan x n_pnt
                em = epoch.get_data()[:,:,win[0]:win[1]].mean(axis=2)
                n_seg, n_chan = em.shape
                if not n_chan == len(channels):
                    print(f'Expected {len(channels)} channels but only see {n_chan}')

                n_obs = n_seg * n_chan
                df = pd.DataFrame({
                    'Subject': [subject] * n_obs,
                    'Channel': channels * n_seg,
                    'Cluster': [clust] * n_obs,
                    'ERP' : [erp] * n_obs,
                    'Support' : [s] * n_obs,
                    'Moral' : [m] * n_obs,
                    'uV' : em.flatten()*1e6
                })

                dfs.append(df) 
    if dfs:
        iss_df = pd.concat(dfs)

    return(iss_df)


def export_social(subject, eeg, events, codes, clust_dict=clust_dict,
                 reject_criteria=reject_criteria, flat_criteria=flat_criteria,
                 bin_moral=False):
    """    
    Calculate the mean voltage for each channel within the EPN, MFN, and LPP 
    time windows, separately for each distinct moral conviction rating.
    
    Parameters
    ----------
    subject : char
        Subject id
    eeg : pandas DataFrame
        subject's eeg from import_subeeg
    events : pandas DataFrame
        subjects event structure from import_subeeg
    codes : list
        all codes for social onsets that exist for this subject
        
    Returns
    -------
    iss_df : pandas DataFrame
        collected data written to file
    """
    erps = ['epn', 'mfn', 'lpp']
    clusters = ['front', 'post']
    soc = []
    sup = []
    mor = []
    event_codes = []
    for code in codes:
        event_codes.append(code)
        ms, _, m = parse_soc(code)
        soc.append(ms)
        mor.append(m)

    sub_df = pd.DataFrame({'stim_id': event_codes, 
                           'social' : soc,
                           'moral' : mor})
    
    if bin_moral:
        cases = [
            (sub_df["moral"].lt(3)),
            (sub_df["moral"].ge(3) & sub_df["moral"].lt(5)),
            (sub_df["moral"] == 5),
            (sub_df["moral"].gt(5) & sub_df["moral"].le(7)),
            (sub_df["moral"].gt(7)),
        ]
        bins = [1, 2, 3, 4, 5]

        sub_df["moral_bin"] = np.select(cases, bins)
        sub_df['moral'] = sub_df['moral_bin']
        
    soc_df = pd.DataFrame(
        columns=['Subject','Channel','Cluster','ERP','SocialSupport','Moral','uV']
    )

    dfs = []

    for clust in clusters:
        channels = clust_dict[clust]
        
        for soc_id in sub_df['social'].unique():
            
            for mc_id in sub_df['moral'].unique():
                
                code_vals = sub_df[ (sub_df['moral'] == mc_id) & (sub_df['social'] == soc_id) ]['stim_id']
                
                if code_vals.empty:
                    pass #skip cause there were none
                else:
                    ms, _, m = parse_soc(code_vals.values[0], bin_moral)
                    epoch = mne.Epochs(eeg, events, event_id=code_vals.tolist(),
                                       tmin=-0.2, tmax=1, picks=channels,
                                       reject=reject_criteria, flat=flat_criteria,
                                       reject_by_annotation=False, preload=True, verbose='WARNING')
                    
                    for erp in erps:
                        win = [ms2pnt(x+bl) for x in time_dict['soc'][erp]]
                        em = epoch.get_data()[:,:,win[0]:win[1]].mean(axis=2)
                        n_seg, n_chan = em.shape
                        if not n_chan == len(channels):
                            print(f'Expected {len(channels)} channels but only see {n_chan}')
                        
                        n_obs = n_seg*n_chan
                        df = pd.DataFrame({
                            'Subject': [subject] * n_obs,
                            'Channel': channels * n_seg,
                            'Cluster': [clust] * n_obs,
                            'ERP' : [erp] * n_obs,
                            'SocialSupport' : [ms] * n_obs,
                            'Moral' : [m] * n_obs,
                            'uV' : em.flatten()*1e6
                            })
                        dfs.append(df) 
    if dfs:
        soc_df = pd.concat(dfs)

    return(soc_df)


def export_pic(subject, eeg, events, codes, clust_dict=clust_dict,
               reject_criteria=reject_criteria, flat_criteria=flat_criteria,
               bin_moral=False):
    """    
    Calculate the mean voltage for each channel within the EPN, MFN, and LPP 
    time windows, separately for each distinct moral conviction rating and 
    protestor position (Agree|Disagree).
    
    Parameters
    ----------
    subject : char
        Subject id
    eeg : pandas DataFrame
        subject's eeg from import_subeeg
    events : pandas DataFrame
        subjects event structure from import_subeeg
    codes : list
        all codes for social onsets that exist for this subject
    (optional)
    bin_moral : bool
        if True, collapse moral conviction ratings to 1, 2, 3, 4, 5
        so that 1=1/2, 2=3/4, 3=5, 4=6/7, 5=8/9. Otherwise, extract
        separately for up to 9 different moral ratings
        (default=False)
        
    Returns
    -------
    iss_df : pandas DataFrame
        collected data written to file
    """
    erps = ['epn', 'mfn', 'lpp']
    clusters = ['front', 'post']
    pos = []
    soc = []
    sup = []
    mor = []
    event_codes = []
    for code in codes:
        event_codes.append(code)
        p, ms, _, m = parse_pic(code)
        pos.append(p)
        sup.append(s)
        mor.append(m)

    sub_df = pd.DataFrame({'stim_id': event_codes, 
                           'position' : pos,
                           'support' : sup, 
                           'moral' : mor})
    
    if bin_moral:
        cases = [
            (sub_df["moral"].lt(3)),
            (sub_df["moral"].ge(3) & sub_df["moral"].lt(5)),
            (sub_df["moral"] == 5),
            (sub_df["moral"].gt(5) & sub_df["moral"].le(7)),
            (sub_df["moral"].gt(7)),
        ]
        bins = [1, 2, 3, 4, 5]

        sub_df["moral_bin"] = np.select(cases, bins)
        sub_df['moral'] = sub_df['moral_bin']
    
    soc_df = pd.DataFrame(
        columns=['Subject','Channel','Cluster','ERP','ProtestPosition','Moral','uV']
    )

    dfs = []

    for clust in clusters:
        channels = clust_dict[clust]
        
        for pos_id in sub_df['position'].unique():
            
            for mc_id in sub_df['moral'].unique():
                
                code_vals = sub_df[ (sub_df['moral'] == mc_id) & (sub_df['position'] == pos_id) ]['stim_id']
                
                if code_vals.empty:
                    pass #skip cause there were none
                else:
                    p, _, _, m = parse_pic(code_vals.values[0], bin_moral)
                    epoch = mne.Epochs(eeg, events, event_id=code_vals.tolist(),
                                       tmin=-0.2, tmax=1, picks=channels,
                                       reject=reject_criteria, flat=flat_criteria,
                                       reject_by_annotation=False, preload=True, verbose='WARNING')
                    
                    for erp in erps:
                        
                        win = [ms2pnt(x+bl) for x in time_dict['pic'][erp]]
                        em = epoch.get_data()[:,:,win[0]:win[1]].mean(axis=2)
                        n_seg, n_chan = em.shape
                        if not n_chan == len(channels):
                            print(f'Expected {len(channels)} channels but only see {n_chan}')
                        
                        n_obs = n_seg*n_chan
                        df = pd.DataFrame({
                            'Subject': [subject] * n_obs,
                            'Channel': channels * n_seg,
                            'Cluster': [clust] * n_obs,
                            'ERP' : [erp] * n_obs,
                            'ProtestPosition' : [p] * n_obs,
                            'Moral' : [m] * n_obs,
                            'uV' : em.flatten()*1e6
                            })
                        dfs.append(df) 
    if dfs:
        soc_df = pd.concat(dfs)

    return(soc_df)
```

```python
def export_social(subject, eeg, events, issue_codes, clust_dict=clust_dict,
                 reject_criteria=reject_criteria, flat_criteria=flat_criteria):
    erps = ['epn', 'mfn', 'lpp']
    clusters = ['front', 'post']
    iss_sup = []
    iss_mor = []
    iss_codes = []
    for iss in issue_codes:
        iss_codes.append(iss)
        s, m = parse_issue(iss)
        iss_sup.append(s)
        iss_mor.append(m)
    
    sub_df_iss = pd.DataFrame({'iss_id': iss_codes, 
                           'support' : iss_sup, 
                           'moral' : iss_mor})
    iss_df = pd.DataFrame(
        columns=['Subject','Channel','Cluster','ERP','Support','Moral','uV']
    )

    dfs = []

    for clust in clusters:
        channels = clust_dict[clust]

        for ui in sub_df_iss['moral'].unique():
            code_vals = sub_df_iss[sub_df_iss['moral'] == ui]['iss_id'].values
            s, m = parse_issue(code_vals[0])
            epoch = mne.Epochs(eeg, events, event_id = code_vals.tolist(), tmin=-0.2, tmax=1,
                               picks = channels, reject=reject_criteria,
                               flat=flat_criteria, reject_by_annotation=False,
                               preload=True, verbose='WARNING')

            for erp in erps:
                #print(f'Now working with erp {erp}')
                win = [ms2pnt(x+bl) for x in time_dict['iss'][erp]]
                # n_seg x n_chan x n_pnt
                em = epoch.get_data()[:,:,win[0]:win[1]].mean(axis=2)
                n_seg, n_chan = em.shape
                if not n_chan == len(channels):
                    print(f'Expected {len(channels)} channels but only see {n_chan}')

                n_obs = n_seg * n_chan
                df = pd.DataFrame({
                    'Subject': [subject] * n_obs,
                    'Channel': channels * n_seg,
                    'Cluster': [clust] * n_obs,
                    'ERP' : [erp] * n_obs,
                    'Support' : [s] * n_obs,
                    'Moral' : [m] * n_obs,
                    'uV' : em.flatten()*1e6
                })

                dfs.append(df) 
    if dfs:
        iss_df = pd.concat(dfs)

    return(iss_df)
```
